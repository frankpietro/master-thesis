```python
def generate_patients(num, max_patients):
    return np.random.randint(1, max_patients+1, size=num)

def generate_visits(num, max_visits, p_distr=[0.5, 0.2, 0.3]):
    return np.random.choice(range(1, max_visits+1), size=num, p=p_distr)

def generate_timeslots(num, num_timeslots, ts_distr_array):
    sample_length = int(num / len(ts_distr_array))
    r = num % len(ts_distr_array)
    # generate samples and concat all samples in one matrix (num, num_timeslots
    first = True
    
    # variable for full matrix
    matrix = None

    for p_distr in ts_distr_array:
        if first:
            first = False
            sample_length += r

        samples = np.random.choice(2, size=(sample_length, num_timeslots), p=p_distr)
        # all samples must have at least one 1 per row
        for i in range(sample_length):
            if not np.any(samples[i]):
                samples[i][np.random.randint(0, num_timeslots)] = 1
                
        if matrix is None:
            matrix = samples
        else:
            matrix = np.concatenate((matrix, samples), axis=0)

    return matrix

def generate_priorities(num, p_distr=[0.8, 0.2]):
    return np.random.choice((1,2), size=num, p=p_distr)


def print_opl(elem):
    print("<{}, {}, {}, {}, {}>,".format(*elem))


def generate_requests(num, max_patients, max_visits, ts_distr_array, num_timeslots=10, p_distr=[0.8, 0.2]):
    patients = generate_patients(num, max_patients)
    visits = generate_visits(num, max_visits)

    timeslots = generate_timeslots(num, num_timeslots, ts_distr_array)
    priorities = generate_priorities(num)
    for elem in list(zip(range(1, num+1), patients, visits, timeslots, priorities)):
        print_opl(elem)
```

```python
# generate a 20x20 matrix of the commutation time among patients
# each element has a 0.2 probability of having a value between 1 and 10 and 0.8 probability of having a value between 11 and 20
# diagonal must be 0
# symmetric matrix
# each element must be an integer
# enforce triangular inequality

def generate_commutation_matrix(size=[27, 27]):
    # generate distances with gaussian distribution
    matrix = np.random.normal(18, 9, size=size)
    # convert to integer
    matrix = matrix.astype(int)
    # convert all negative numbers to 1
    matrix[matrix < 1] = 1
    # diagonal must be 0
    np.fill_diagonal(matrix, 0)
    # symmetric matrix
    matrix = np.triu(matrix) + np.triu(matrix, 1).T
    return matrix

comm_matrix = generate_commutation_matrix()
comm_matrix
```

```python
def generate_requests(num_patients, num_days, num_timeslots, num_visits, prob=0.5):
    requests = []
    index = 1
    for patient in range(1, num_patients+1):
        for day in range(1, num_days+1):
            # random variable between 0 and 1
            if np.random.rand() > prob:
                timeslot = np.random.randint(1, num_timeslots+1)
                visit = np.random.choice(range(1, num_visits + 1), p=[0.55, 0.35, 0.1])
                requests.append((index, patient, visit, day, timeslot))
                index += 1

    return requests
```
