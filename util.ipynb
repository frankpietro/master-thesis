```python
# generate an array of dimensions 20 times 10 times 3, with all values set to 0
# each 3-dimensional element having a 0.2 probability of having one number to 1 and the others to 0 and 0.8 probability of having all zeros

import numpy as np


def generate_array(arr, size, p_distr):
    return np.random.choice(len(arr), size=size, p=p_distr)
    
arr = [[0,0,0], [1,0,0], [0,1,0], [0,0,1]]
size = (20,10)
p_distr = [0.6, 0.15, 0.15, 0.1]
```

```python
indexes = generate_array(arr, size, p_distr)
```

```python
# generate 3-dimensional matrix by substituting the indexes with the values
matrix = np.array(arr)[indexes]
```
